```sql
USE AdventureWorks2019
```

```sql
--Declarar Variável--
DECLARE
    @nome varchar(60),
    @idade INT
--Atribuição de Dados--]
BEGIN
    SET @nome = 'Tiago'
    SET @idade=18
    SELECT @nome = 'Tiago'
    SELECT @nome = FirstName FROM Person.Person WHERE BusinessEntityID=1000
    IF @idade< 18 
        PRINT('Menor de Idade')
    ELSE
        PRINT('Maior de Idade')
END
```

```sql
   --Declarar Variável--
DECLARE
    @nome varchar(50),
    @tipoPesquisa INT
--Atribuição de Dados--]
BEGIN
    SET @nome = 'Adam'
    SET @tipoPesquisa = 1
     IF @tipoPesquisa = 1
    BEGIN 
        PRINT('Pesquisa por Nome')
        SELECT FirstName, LastName FROM Person.Person WHERE FirstName = @nome
    END
    ELSE
    BEGIN   
        PRINT('Pesquisa por Sobrenome')
        SELECT FirstName, LastName FROM Person.Person WHERE LastName = @nome
    END
END     
```

```sql
DECLARE 
    @idade INT,
    @categoria VARCHAR(10)

BEGIN
    SET @idade=12

    SET @categoria = CASE @idade
                        WHEN 10 THEN  'Quinto'
                        WHEN 11 THEN  'Sexto'
                        WHEN 12 THEN  'Sétimo'
                        WHEN 13 THEN  'Oitavo'
                        WHEN 14 THEN  'Nono'
                    END
    PRINT(@categoria)
    SET @categoria = CASE
                        WHEN @idade> 18 THEN 'Adulto'
                        WHEN @idade> 15 THEN 'Juvenil'
                        WHEN @idade> 10 THEN 'Infanto Juvenil'
                        ELSE 'Infantil'
                   END
    PRINT(@categoria)
END
```

```sql
DECLARE @count INT
BEGIN
    SET @count=0
    WHILE @count < 10
    BEGIN
        Set @count=@count+1
        IF @count = 5
          BREAK
        PRINT(@count)
    END
END
```

```sql

DECLARE @count INT
BEGIN
    SET @count=0
    WHILE @count < 10
    BEGIN
        PRINT(@count)
        Set @count=@count+1
        IF @count = 5
          CONTINUE
        PRINT(@count)
    END
END
```

```sql
CREATE PROCEDURE usp_Teste @a INT, @b INT
AS
BEGIN
    DECLARE
    @res INT
    SET @res = @a + @b
    PRINT('Soma:' + CAST(@res as VARCHAR))
END
```

```sql
EXEC usp_teste 2,3
```

```sql
ALTER PROCEDURE usp_Teste @a INT, @b INT
AS
BEGIN
    DECLARE
    @res INT
    SET @res = @a + @b
    PRINT('Soma:' + CAST(@res as VARCHAR))
END
```

```sql
EXEC usp_Teste 2,3
```

```sql
CREATE or ALTER PROCEDURE usp_soma(@a int, @b int, @res int OUTPUT)
AS
BEGIN
    SET @res = @a + @b
END
```

```sql
DECLARE @r INT
EXEC usp_soma @b=2, @a=3, @res=@r output
print('Soma: ' + Cast(@r as varchar))
```
